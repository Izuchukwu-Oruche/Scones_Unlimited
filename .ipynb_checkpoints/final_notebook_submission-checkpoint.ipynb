### 1. Extract the data from the hosting service

In the cell below, a function `extract_cifar_data` will be defined that extracts the python version of the CIFAR-100 dataset. The CIFAR dataaset is open source and generously hosted by the University of Toronto at: https://www.cs.toronto.edu/~kriz/cifar-100-python.tar.gz 

````python
import requests

def extract_cifar_data(url, filename="cifar.tar.gz"):
    """A function for extracting the CIFAR-100 dataset and storing it as a gzipped file
    
    Arguments:
    url      -- the URL where the dataset is hosted
    filename -- the full path where the dataset will be written
    
    """
    
    # Requesting data from data URL
    
    with requests.get(url, stream=True) as r:
        r.raise_for_status()  
        with open(filename, "wb") as file_context:
            for chunk in r.iter_content(chunk_size=8192):
                file_context.write(chunk)
    return
````

````python
# Extraction of the dataset

extract_cifar_data("https://www.cs.toronto.edu/~kriz/cifar-100-python.tar.gz")    
````

For a simple gut-check, one of the images will be transformed. Each 1024 items in a row is a channel (red, green, then blue). Each 32 items in the channel are a row in the 32x32 image. Using python, these channels can be stacked into a 32x32x3 array, and saved as a PNG file:

````python
import numpy as np

# Each 1024 in a row is a channel (red, green, then blue)
row = dataset_train[b'data'][0]
red, green, blue = row[0:1024], row[1024:2048], row[2048:]

# Each 32 items in the channel are a row in the 32x32 image
red = red.reshape(32,32)
green = green.reshape(32,32)
blue = blue.reshape(32,32)

# Combining the channels into a 32x32x3 image!
combined = np.dstack((red,green,blue))
````

Only objects with labels 8 and 48 are needed. Below a dataset is constructed and I'll drop all rows that aren't Bicycles or Motorcycles.

````python
#Constructing the train dataframe
df_train = pd.DataFrame({
    "filenames": dataset_train[b'filenames'],
    "labels": dataset_train[b'fine_labels'],
    "row": range(len(dataset_train[b'filenames']))
})
wanted_labels = [8, 48]

# Dropping all rows from df_train where label is not 8 or 48
df_train = df_train[df_train["labels"].isin(wanted_labels)]



# Decoding df_train.filenames so they are regular strings
df_train["filenames"] = df_train["filenames"].apply(
    lambda x: x.decode("utf-8")
)

# Constructing the test dataframe
df_test = pd.DataFrame({
    "filenames": dataset_test[b'filenames'],
    "labels": dataset_test[b'fine_labels'],
    "row": range(len(dataset_test[b'filenames']))
})

# Dropping all rows from df_test where label is not 8 or 48
df_test = df_test = df_test[df_test["labels"].isin(wanted_labels)]

# Decoding df_test.filenames so they are regular strings
df_test["filenames"] = df_test["filenames"].apply(
    lambda x: x.decode("utf-8")
)
````

These will be composed together into a function that will save all the images into the `./test` and `./train` directories. 


````python
def save_images(image, folder, n):
    #Grabbing the image data in row-major form
    
    # Consolidated stacking/reshaping from earlier
    target = np.dstack((
    image[0:1024].reshape(32,32),
    image[1024:2048].reshape(32,32),
    image[2048:].reshape(32,32)
    ))
    
    if folder == "train":
        image_name = dataset_train[b'filenames'][n].decode("utf-8")
        
    elif folder == "test":
        image_name = dataset_test[b'filenames'][n].decode("utf-8")
        
  
    
    plt.imsave(folder+"/"+image_name, target)
    
    
    # Returning any signal data you want for debugging
    return None

# saving all images using the save_images function
n = 0
for image in dataset_train[b'data']:
    if dataset_train[b'fine_labels'][n] == 8 or dataset_train[b'fine_labels'][n] == 48:
        save_images(image, "train", n)
    n += 1
n = 0    
for image in dataset_test[b'data']:
    if dataset_test[b'fine_labels'][n] == 8 or dataset_test[b'fine_labels'][n] == 48:
        save_images(image, "test", n)
    n += 1
````

## Model Training

For Image Classification, Sagemaker [also expects metadata](https://docs.aws.amazon.com/sagemaker/latest/dg/image-classification.html) e.g. in the form of TSV files with labels and filepaths. These can be generated using the pandas Dataframe from earlier

````python
def to_metadata_file(df, prefix):
    """A function for creating and saving manifest files using dataframes
    
    Arguments:
    df      -- the dataframe to be used
    prefix -- for the file name, if it's a train or test manifest file
    
    """
    df["s3_path"] = df["filenames"]
    df["labels"] = df["labels"].apply(lambda x: 0 if x==8 else 1)
    return df[["row", "labels", "s3_path"]].to_csv(
        f"{prefix}.lst", sep="\t", index=False, header=False
    )
    
to_metadata_file(df_train.copy(), "train")
to_metadata_file(df_test.copy(), "test")
````

The manifest files will be uploaded to S3.

````python
import boto3

# Uploading manifest files
boto3.Session().resource('s3').Bucket(
    bucket).Object('train.lst').upload_file('./train.lst')
boto3.Session().resource('s3').Bucket(
    bucket).Object('test.lst').upload_file('./test.lst')
````

### Draft Lambdas and Step Function Workflow

The operations team uses Step Functions to orchestrate serverless workflows. One of the nice things about Step Functions is that [workflows can call other workflows](https://docs.aws.amazon.com/step-functions/latest/dg/connect-stepfunctions.html), so the team can easily plug the workflow into the broader production architecture for Scones Unlimited.

In this next stage three lambda functions will be written and deployed. Then the Step Functions visual editor will be used to chain them together. The functions are going to work with a simple data object.

```python
{
    "inferences": [], # Output of predictor.predict
    "s3_key": "", # Source data S3 key
    "s3_bucket": "", # Source data S3 bucket
    "image_data": ""  # base64 encoded string containing the image data
}
```


Using these fields, the functions will read and write the necessary data to execute the workflow. The first Lambda function will copy an object from S3, base64 encode it, and then return it to the step function as `image_data` in an event.

Go to the Lambda dashboard and create a new Lambda function with a descriptive name like "serializeImageData" and select thr 'Python 3.8' runtime. Add the same permissions as the SageMaker role you created earlier. (Reminder: you do this in the Configuration tab under "Permissions"). Once you're ready, use the starter code below to craft your Lambda handler:

In the Sagemaker Console, the first lambda function was created with a "Python 3.8" runtime and with the same Sagemaker role. It's lambda handler code is below: 

```python
import json
import boto3
import base64
import sagemaker
from sagemaker.serializers import IdentitySerializer
from sagemaker.predictor import Predictor


s3 = boto3.client('s3')

def lambda_handler(event, context):
    """A function to serialize target data from S3"""
    
    # Get the s3 address from the Step Function event input
    key = event['s3_key']
    bucket = event['s3_bucket']
    
    # Download the data from s3 to /tmp/image.png
    s3.download_file(Bucket=bucket, Key=key, Filename="/tmp/image.png")
    
    # We read the data from a file and encode it
    with open("/tmp/image.png", "rb") as f:
        image_data = base64.b64encode(f.read()) # decode to ensure it's a string

    # Pass the data back to the Step Function
    print("Event:", event.keys())
    return {
        'statusCode': 200,
        'body': { # Ensure the body is properly JSON encoded
            "image_data": image_data,
            "s3_bucket": bucket,
            "s3_key": key,
            "inferences": []
        }
    }
```

The next function is responsible for the classification part - the image output from the previous function will be taken, decoded , and its inferences passed back to the Step Function.

Because this Lambda will have runtime dependencies (i.e. the SageMaker SDK) you'll need to package them in your function. 



```python
import json
import boto3
import base64
import sagemaker
from sagemaker.serializers import IdentitySerializer
from sagemaker.predictor import Predictor


# Fill this in with the name of your deployed model
ENDPOINT = "image-classification-2024-03-21-19-20-42-627"

def lambda_handler(event, context):

    # Decode the image data
    image = base64.b64decode(event["image_data"])

    # Instantiate a Predictor
    predictor = Predictor(ENDPOINT)

    # For this model the IdentitySerializer needs to be "image/png"
    predictor.serializer = IdentitySerializer("image/png")
    
    # Make a prediction:
    inferences = predictor.predict(image)
    
    # Return the data back to the step function  
    event["inferences"] = inferences.decode('utf-8')
    return {
        'statusCode': 200,
        'body': json.dumps(event)
    }


```

Finally low-confidence predictions will be filtered. A threshold of 0.9 will be set as the filter.

```python
import json


THRESHOLD = .90


def lambda_handler(event, context):
    
    # Grab the inferences from the event
    inferences = event["inferences"]
    
    inferences = eval(inferences)
    
    # Check if any values in our inferences are above THRESHOLD
    meets_threshold = any(x > THRESHOLD for x in inferences)
    
    # If our threshold is met, pass our data back out of the
    # Step Function, else, end the Step Function with an error
    if meets_threshold:
        pass
    else:
        raise("THRESHOLD_CONFIDENCE_NOT_MET")

    return {
        'statusCode': 200,
        'body': json.dumps(event)
    }
```

````python
import base64
import io
import matplotlib.pyplot as plt
from PIL import Image
import pandas as pd

def plot_inference_details(image_data_base64, confidence_scores, inference_time):
    # Decode the base64 image data
    image_bytes = base64.b64decode(image_data_base64)
    image = Image.open(io.BytesIO(image_bytes))
    
    # Prepare data for plotting
    categories = ['Bicycle', 'Motorcycle']  # Example category labels
    scores_df = pd.DataFrame({
        'Category': categories,
        'Confidence': confidence_scores
    })
    
    # Create a figure to hold the image and the bar chart
    fig, ax = plt.subplots(1, 2, figsize=(12, 5))
    
    # Plot the image
    ax[0].imshow(image)
    ax[0].axis('off')  # Hide axes for the image
    
    # Plot the bar chart for confidence scores
    scores_df.plot(kind='bar', x='Category', y='Confidence', legend=None, ax=ax[1])
    ax[1].set_ylim(0, 1)
    ax[1].set_ylabel('Confidence')
    ax[1].set_title('Inference Confidence Scores')
    
    # Display inference time
    plt.suptitle(f'Inference Time: {inference_time}')
    
    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()



# Plot the details for this example

def complex_getter(obj):
    confidence_scores = obj["captureData"]["endpointOutput"]["data"]
    inference_time = obj["eventMetadata"]["inferenceTime"]
    image_data_base_64 = obj["captureData"]["endpointInput"]["data"]
    return image_data_base_64, json.loads(confidence_scores), inference_time

for inference in json_data:
    image_data_base_64, confidence_scores, inference_time = complex_getter(inference)
    plot_inference_details(image_data_base_64, confidence_scores, inference_time)
````
